```python
import matplotlib.pyplot as plt                               
import pandas as pd
                      
```

**Đọc dữ liệu từ drive**

```python
data = pd.read_csv('/content/drive/MyDrive/machineLearning/score_CS/wecode.csv')
```

```python
new_data = data
```

```python
new_data
```

**Gom nhóm các bài của cùng 1 user_id**

```python
group = new_data.groupby(by='user_id')
new_data2 = pd.DataFrame({'assignment_id': group['assignment_id'].nunique(),
                          'is_final': group['is_final'].sum(),
                           'pre_score':group['pre_score'].sum(),
                          })
 
```

```python
new_data2
```

```python
score
```

**Vì type của CK là object không vẽ được plt nên đổi thành numeric**

```python
score['CK'] = score['CK'].apply(pd.to_numeric, errors ='coerce' )
```

```python
temp = pd.DataFrame({'user_id':score['id'], 'CK':score['CK']})
```

```python
new_data3 = pd.merge(new_data2,temp, on='user_id')
```

```python
new_data3
```

**Vẽ plt dựa trên tổng pre_score và CK**

```python
plt.scatter(new_data3[['pre_score']], new_data3[['CK']],  color='black')

plt.show()
```

**Vẽ plt dựa trên tổng is_final và CK**

```python
plt.scatter(new_data3[['is_final']], new_data3[['CK']],  color='purple')

plt.show()
```
